```python
import matplotlib.pyplot as plt
import pandas as pd
import scienceplots
import torch
from tqdm import tqdm

from core.environment.env import Environment
from core.model.decision_network.network import DecisionNetwork
```

```python
plt.style.use(['science', 'grid'])
```

```python
model = DecisionNetwork()
model.load_state_dict(torch.load('./policy_net.pth'))
model = model.eval()
```

```python
env = Environment(train=False)
```

```python
state, info = env.reset()

rewards = []
actions = []
infos = []

for i in tqdm(range(1000)):
    state = torch.from_numpy(state).type(torch.float32).unsqueeze(0)
    action = model(state)
    action = action.argmax(dim=1).item()
    observation, reward, terminated, truncated, info = env.step(action)
    state = observation
    
    rewards.append(reward)
    actions.append(action)
    infos.append(info)
    
    if terminated:
        break
    
```
